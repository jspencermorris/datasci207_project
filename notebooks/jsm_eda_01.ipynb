```python
import pandas as pd
```

```python
# load tabular data
# NOTE: there is one more dataset (original galaxy zoo 1) that also includes overlapping galaxy classifications, with labels spiral, elliptital, irregular...
filename_mapping = pd.read_csv('../data/raw/gz2/gz2_filename_mapping.csv', header=0) # galaxies - includes mapping definition between image filenames and (dr7)objid
hart16 = pd.read_csv('../data/raw/gz2/gz2_hart16.csv', header=0) # galaxies - includes detailed (sub)classification labels using dr7objid
dr16_galaxy = pd.read_csv('../data/raw/skyserver_sdss/dr16_galaxySpecPhoto.csv', header=0) # galaxies - includes (super)classification labels, spectral data, and photo data
dr16_other = pd.read_csv('../data/raw/skyserver_sdss/dr16_otherSpecPhoto.csv', header=0)  # non-galaxy objects - includes (super)classification labels, spectral data, and photo data
```

```python
# rename identifier columns for improved clarity
filename_mapping.rename(columns={'objid':'dr7objid'}, inplace=True)
```

```python
# display data heads
display(filename_mapping.head(3))
display(hart16.head(3)) # main features: [];  main labels: [gz2_class]
display(dr16_galaxy.head(3)) # main features: [clean, u, g, r, i, z];  main labels: [p_el, p_cw, p_acw, p_edge, p_dk, p_mg, p_cs, sdss_class_number, sdss_class_string, subClass, class_noqso]
display(dr16_other.head(3)) # main features: [clean, u, g, r, i, z];  main labels: [p_el, p_cw, p_acw, p_edge, p_dk, p_mg, p_cs, sdss_class_number, sdss_class_string, subClass, class_noqso]
```

```python
# display dataframe shapes
display(filename_mapping.shape)
display(hart16.shape)
display(dr16_galaxy.shape)
display(dr16_other.shape)
```

```python
# determine number of unique labels

print(len(hart16['gz2_class'].unique())) # each class is a galaxy morphology subclass

print(len(dr16_galaxy['sdss_class_number'].unique())) # 3=galaxy, 6=star
print(len(dr16_galaxy['sdss_class_string'].unique())) # galaxy, star, qso
print(dr16_galaxy['sdss_class_string'].unique()) # print class labels for galaxy set
```

```python
# evaluate the number of overlapping records in different tables

# first and second tables
merged_df = pd.merge(filename_mapping, hart16, left_on='dr7objid', right_on='dr7objid', how='inner')
print(f"Matching values: {len(merged_df['dr7objid'].tolist())}")
print(f"Percentage of matching values in df1: {(len(merged_df) / len(filename_mapping)) * 100}%")
print(f"Percentage of matching values in df2: {(len(merged_df) / len(hart16)) * 100}%")

# first and third tables
merged_df = pd.merge(filename_mapping, dr16_galaxy, left_on='dr7objid', right_on='dr7objid', how='inner')
print(f"Matching values: {len(merged_df['dr7objid'].tolist())}")
print(f"Percentage of matching values in df1: {(len(merged_df) / len(filename_mapping)) * 100}%")
print(f"Percentage of matching values in df2: {(len(merged_df) / len(dr16_galaxy)) * 100}%")

# second and third tables
merged_df = pd.merge(hart16, dr16_galaxy, left_on='dr7objid', right_on='dr7objid', how='inner')
print(f"Matching values: {len(merged_df['dr7objid'].tolist())}")
print(f"Percentage of matching values in df1: {(len(merged_df) / len(hart16)) * 100}%")
print(f"Percentage of matching values in df2: {(len(merged_df) / len(dr16_galaxy)) * 100}%")

# first and fourth tables

# second and fourth tables

# third and fourth tables
merged_df = pd.merge(dr16_galaxy, dr16_other['dr8objid'], left_on='dr8objid', right_on='dr8objid', how='inner')
print(f"Matching values: {len(merged_df['dr8objid'].tolist())}")
print(f"Percentage of matching values in df1: {(len(merged_df) / len(dr16_galaxy)) * 100}%")
print(f"Percentage of matching values in df2: {(len(merged_df) / len(dr16_other)) * 100}%")
```
